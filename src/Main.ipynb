### Part 1: The Travelling Robot Problem
The code below is in regards to the first part of the assignment: _solving TSP using GA._ The expected graph of `minimum distance / generations` graph is a slowing decrease. The hyperparameters are tuned specifically for the intention.

```python
import time
from GeneticAlgorithm import GeneticAlgorithm
from TSPData import TSPData
import matplotlib.pyplot as plt

# Please keep your parameters for the Genetic Algorithm easily changeable here
population_size = 100
generations = 1000
crossover_prob = 0.5
mutation_prob = 0.005
persist_file = "./../data/optimal_tsp"

# Setup optimization
tsp_data = TSPData.read_from_file(persist_file)
ga = GeneticAlgorithm(generations, population_size, crossover_prob, mutation_prob)

# Run optimzation and write to file
solution = ga.solve_tsp(tsp_data)
plt.xlabel("Generation")
plt.ylabel("Minimum distance")
plt.plot(ga.values)

tsp_data.write_action_file(solution, "./../data/tsp_solution.txt")
```

### Part 2: Path Finding Through Ant Colony Optimization

It turns out that for each maze, parameters need different tunings. For that reason codeblocks below are seperated for EASY, MEDIUM, and HARD mazes.

<h1>Easy

```python
# Please keep your parameters for the ACO easily changeable here TODO
gen = 10
no_gen = 35
q = 5
evap = 0.1

# Construct the optimization objects
maze = Maze.create_maze("./../data/easy_maze.txt")
spec = PathSpecification.read_coordinates("./../data/easy_coordinates.txt")
aco = AntColonyOptimization(maze, gen, no_gen, q, evap)

# Save starting time
start_time = int(round(time.time() * 1000))

# Run optimization
shortest_route = aco.find_shortest_route(spec)

# Print time taken
print("Time taken: " + str((int(round(time.time() * 1000)) - start_time) / 1000.0))

# Save solution
shortest_route.write_to_file("./../data/62_easy.txt")

# Print route size
print("Route size: " + str(shortest_route.size()))
```

```python
plt.plot(aco.gen_averages)
plt.xlabel("Generations")
plt.ylabel("Average route distance")
plt.title("Avg route distance in each generation, easy maze")
plt.show()
```
